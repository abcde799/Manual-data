```python
#Here we scrape the locally saved webpages from the relevant website using Beautiful soup along with the lxml parser. 
#This was initially done for the first page, then we gathered the information into a pandas dataframe by
#looping over the webpages. Saving the webpages is covered in another notebook.

from bs4 import BeautifulSoup
```

```python
#open and/or read html from a local file

def open_html(path):
    with open(path, 'rb') as f:
        return f.read()
    
    
```

```python
#Create list of soup objects

soup_list = []

for i in range(1,371):
          
    x = open_html('manual'+str(i)+'_com')

    soup = BeautifulSoup(x, 'lxml')
    
    soup_list.append(soup)
    
    
    
    

```

```python
print(len(soup_list))
```

```python
#year records

all_year_records = []

for i in range(0,370):
    
    year_vehicle_location = soup_list[i].find_all('div', attrs={'class':'img-container img-container-block hidden-sm-up'})
    
    year_records = []

    for index in range(len(year_vehicle_location)): 

        year = year_vehicle_location[index].img['alt'][0:4]
    
        year_records.append(year)

    all_year_records.append(year_records)        
 

    
    

```

```python
#Turn list of lists into a single list 

def flat(list_of_lists):

    flattened = [val for sublist in list_of_lists for val in sublist]

    return flattened

all_year_records = flat(all_year_records)

len(all_year_records)



```

```python
#model records

all_model_records = []

for i in range(0,370):
    
    year_vehicle_location = soup_list[i].find_all('div', attrs={'class':'img-container img-container-block hidden-sm-up'})
    
    model_records = []

    for index in range(len(year_vehicle_location)): 

        model = year_vehicle_location[index].img['alt'][4:].split('for Sale in')[0]
    
        model_records.append(model)

    all_model_records.append(model_records)    
    

    
    

   
```

```python
all_model_records = flat(all_model_records)

len(all_model_records)
```

```python
#location records

all_location_records = []

for i in range(0,370):
    
    year_vehicle_location = soup_list[i].find_all('div', attrs={'class':'img-container img-container-block hidden-sm-up'})
    
    location_records = []

    for index in range(len(year_vehicle_location)): 

        location = year_vehicle_location[index].img['alt'][4:].split('for Sale in')[1]
    
        location_records.append(location)

    all_location_records.append(location_records)  

```

```python
all_location_records = flat(all_location_records)

len(all_location_records)
```

```python
#mileage records

all_mileage_records = []

for i in range(0,370):
    
    mileage = soup_list[i].find_all('div', attrs={'data-qa':'vehicle-listing-mileage', 'class':'text-truncate', 'data-test':'vehicle-listing-mileage'})
    
    mileage_records = []

    for index in range(len(mileage)):

        miles = mileage[index].contents[2].text
    
        mileage_records.append(miles)

    all_mileage_records.append(mileage_records)  
```

```python
all_mileage_records = flat(all_mileage_records)

print(len(all_location_records))

all_mileage_records[0]
```

```python
#Color records

all_color_records = []

for i in range(0,370):
    
    color = soup_list[i].find_all('div', attrs={'class':'text-truncate', 'data-qa':'vehicle-listing-exterior-colors', 'data-test':'vehicleListingExteriorColors'})
    
    color_records = []

    for index in range(len(color)):

        color_type = color[index].contents[2]
    
        color_records.append(color_type) 

    all_color_records.append(color_records) 
```

```python
all_color_records = flat(all_color_records)

print(len(all_color_records))

all_color_records[0]
```

```python
#Price records

all_price_records = []

for i in range(0,370):
    
    price = soup_list[i].find_all('h4', attrs={'class':'heading-2'})
    
    price_records = []

    for index in range(len(price)):

        cost = price[index].contents[0]
    
        price_records.append(cost)  

    all_price_records.append(price_records) 
```

```python
   
all_price_records = flat(all_price_records)

print(len(all_price_records))

all_price_records[0]   
 
    
  
```

```python
#The big pandas dataframe


import pandas as pd

df = pd.DataFrame({'Model':all_model_records, 'Year': all_year_records, 'Price': all_price_records, 'Location': all_location_records,\
                   'Mileage': all_mileage_records, 'Color': all_color_records})

print(df.info())

df.head(100)
```

```python
#Export to csv file which appears in current directory

df.to_csv('true_car_manual_transmission_07_2_2019', index = False, encoding = 'utf-8')
```
